# Statistics representations

## Import python libraries

```python
import pandas as pd # work with data frames

import matplotlib.pyplot as plt # visual representation
import seaborn as sns # son of the above, but easier, kind of
```

```python
df = pd.read_excel('/content/updated collagen degradation marker to Alberto.xlsx', sheet_name='OI data only') # reading the Excel file
df # showing the contents
```

## Preprocesing

```python
# Add a new column for IPD data
df['ipd_data'] = 0 # Aggregated data
df.loc[df.n == 1, 'ipd_data'] = 1 # IPD data
```

# IPD VS AGG by marker

```python
# Define type of data in each column :)
df = df.astype({"Sex": str, "com mean": float})
df.dtypes
```

```python
# Unique values for Sex column
df.Sex.unique()
```

```python
# Unique values for marker column
df.marker.unique()
```

```python
# Replace of irregular values
df.loc[df.marker == 'sCTX ', 'marker'] = 'sCTX'
df.loc[df.marker == 'sCTX1', 'marker'] = 'sCTX'
df.marker.unique()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm


sm.qqplot(df["com mean"])
plt.show()
```

## Grouping data by marker

```python
gp = df.groupby('marker')
```

```python
# Iteration example with a smile
for name, group in gp:
    print(name) # name of the marker
    print(group) # table of records (articles) that belong to the marker
print("🙂")
```

```python
# Iteration based on the example above
for name, group in gp:
    plt.figure() # create a figure
    plt.title(name) # place the marker name as the title of the figure
    sns.swarmplot( # this creates a swarm plot
        data = group, # based on this table (that comes from the gruops created before)
        x ='ipd_data', # this is the name of the x column
        y ='com mean', # the same for y axis
        hue="Sex" # colour points by some column
    ) 
    sns.violinplot(data = group, x = "ipd_data", y = "com mean", color='gray', inner="quartile")
    plt.show() # this shows the figure and "ends" the figure environment
               # this line and the 3rd makes an environment for figures
               # so we can combine multiple plots into a same figure
    sm.qqplot(group["com mean"],  fit=True, line="45")
    plt.show()
```
