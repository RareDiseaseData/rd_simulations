## Mini example of magic

```python
def simulate_row(r, gender):
  if gender == 'x':  
    new_gender = 'm' if np.random.randint(2, size=1)[0] == 0 else 'f' # male or female
  else:
    new_gender = gender
  new_severity_bag = []
  if 'OI	' in r.severity:
    new_severity = 'not_specified'
  else:
    if 'I' in r.severity:
      new_severity_bag.append('I')
    if 'III' in r.severity:
      new_severity_bag.append('III')
    if 'IV' in r.severity:
      new_severity_bag.append('IV')
    if 'V' in r.severity:
      new_severity_bag.append('V')
    if 'mild' in r.severity:
      new_severity_bag.append('I')
      new_severity_bag.append('III')
      new_severity_bag.append('IV')
      new_severity_bag.append('V')
    new_severity = np.random.choice(new_severity_bag, 1)[0]
  
  new_row = [r['author'], new_gender, new_severity, r.marker, r.age, np.random.normal(r['mean'], r['sd'], 1)[0]]
  return new_row
```

```python
df_new = []

for i, r in df_sel.iterrows(): # iterating rows
  # i: id
  # r: row values
  if r.n > 1:
    if (r.male == 'x'):
      for j in range(int(r.n)):
        new_row = simulate_row(r, 'x')
        df_new.append(new_row)
      # something?
    else: 
      for j in range(int(r.male)):
        new_row = simulate_row(r, 'm')
        df_new.append(new_row)
      for j in range(int(r.female)):
        new_row = simulate_row(r, 'f')
        df_new.append(new_row)

pd.DataFrame(df_new)
```

# Preprocessing

```python
df_sel.isna().sum()
```

```python
df_sel.loc[df_sel.female.isna(), 'female'] = 0
```

```python
df_sel
```

```python
# PENDING!!! 
# Filling n = NA with 1
df_sel.loc[df_sel.n.isna(), 'n'] = 1
df_sel.isna().sum()
```

```python
# Removing Maroteaux from method
df_sel = df_sel.dropna() # equivalent df_sel.dropna(inplace = True)
df_sel.isna().sum()
```

```python
# Convert disease_severity and method values into integers
df_sel.ds = df_sel.ds.astype(int)
df_sel.method = df_sel.method.astype(int)
```

```python
df_sel
```

```python
# One Hot Encoding
df_ds_dummies = pd.get_dummies(df_sel.ds, prefix='ds')
df_method_dummies = pd.get_dummies(df_sel.method, prefix='method')
```

```python
df_sel.join(df_ds_dummies).join(df_method_dummies)
```
